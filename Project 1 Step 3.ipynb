```python
# Part 3 

# Examines data for student motives
cols = ['enrolled_course', 'motive_computing_job',
       'motive_transfer_to_bachelor', 'motive_HS_degree',
       'motive_career_advancement', 'motive_career_change',
       'motive_prof_development', 'motive_job_displacement',
       'motive_Relocation', 'motive_keep_current_with_tech', 'motive_IT_cert',
       'motive_financial', 'motive_personal_enrichment', 'Race/Eth',
       'enrolled_degree']
for i in range(1,13): #looks at motivation columns
    col = cols[i]
    l = df2[col].value_counts()
    #print(df2[col].value_counts())
    print(l)
    print('NaNs: ', df2[col].isna().sum())
    
    print()
```

### What computing majors are students most interested in, and what majors are they not interested in?
#### Both surveys show that Non-Major students are generally interested in taking new computing classes. Majors are not. (Plot 4: 1 & 2)
#### The non majors survey shows that non-major students are mostly taking CMP 135, a basic overview of computing concepts. (Plot 3)
#### The non majors survey shows that non-major students are most interested in cybersecurity, but are also very interested in web development, app development, and game development. (Plot 2)
#### The majors survey shows that by far the most popular class is CMP 128, Computer Science I.
#### With this data, we can see that students are mostly interested in standard computer science & programming classes and cybersecurity, showing an overall interest in degrees and certificates like Computer Science, Information Technology, and Information Security over other degrees and certificates like Data Science and Electronics.

```python
# Looking at the latest survey results (2023) from both the Majors and Non-Majors surveys, 
# what are the race/ethnicity demographic differences between the Majors and Non-Majors respondents?

pd.value_counts(df2['Race/Eth']).plot(kind='bar', color='red', label='df2')

pd.value_counts(nmdf['race/eth']).plot(kind='bar', color='black', label='nmdf')

plt.xlabel('Race / Ethnicity')
plt.ylabel('Count')
plt.title('Race/Ethnicity Comparison between Major and Non-Major Classes')
plt.suptitle('Blue is Majors, Black is Non-Majors', y = .8) 
plt.show()
```

###  Looking at the latest survey results (2023) from both the Majors and Non-Majors surveys, what are the race/ethnicity demographic differences between the Majors and Non-Majors respondents?
#### The proportions of race in majors and non-majors is approximately the same, following the trend that White/Caucasian students are most common, followed by Hispanic/Latino, Choose Not To Reply, and Asian. Also, the dispararity between the number of white respondents and asian respondents is much more drastic in the Major classes compared to the Non-Major classes.
